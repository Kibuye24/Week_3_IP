#DATA CLEANING AND ANALYSIS OF THE TELCOM DATASETS

```python
##View our first telecom dataset
url = "http://bit.ly/Telcom_dataset1"
df_1 = pd.read_csv(url)
df_1.head()
##The three tables will need to be merged. For that, the columns should be same and DATETIME will be changed to DATE_TIME like other datasets
df_1 = df_1.rename(columns = {'DATETIME':'DATE_TIME', 'PRODUTC':'PRODUCT'})
df_1.head()
```

```python
##viewing our second telecom dataset
url = "http://bit.ly/Telcom_dataset2"
df_2 = pd.read_csv(url)
df_2.head()

##The three tables will need to be merged. For that, the columns should be same and DW_A_NUMBER will be changed to DW_A_NUMBER_INT like other datasets
df_2 = df_2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'})
df_2.head()
```

```python
##Viewing our third telecom dataset
url = "http://bit.ly/Telcom_dataset3"
df_3 = pd.read_csv(url)
df_3.head()
##Renaming of the columns siet_ID and CELLID is done to make them in line with the description
df_3 = df_3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID':'SITE_ID'})
df_3.head()
```

```python
##Now we can merge the three telecom datasets to do a uniform data analysis and cleaning
df_telcom = pd.concat([df_1, df_2, df_3], axis=0)
df_telcom
```

```python
##We will then drop the unecessary columns as per the description. Which are the country columns and cell on site.
df_telcom.drop(['DW_A_NUMBER_INT','DW_B_NUMBER_INT','CELL_ON_SITE'], axis = 1, inplace = True) 
##Once the tables have merged and unused columns dropped, we can check for duplicated items and drop the duplicates
df_telcom.duplicated()
##There are some duplicates and we will now drop them
df_telcom = df_telcom.drop_duplicates()
df_telcom
##We also need to check and  drop rows that only have null values
df_telcom.dropna(axis=1, how='all')
##With that, we can conclude that the telcom dataset has been cleaned and is ready for analysis
```

```python
##Here, we want to find the value of the three different products to establish what product the company should focus on
df_telcom['VALUE'].groupby(df_telcom['PRODUCT']).describe()
```

```python
##We will also analyze the time when the various products are most used to determine what time of the day peaks are.
df_telcom['DATE_TIME'].groupby(df_telcom['PRODUCT']).describe()
```

```python
##We will also look at the description of time when people use the products in light of the best time for infrastracture changes to be done
df_telcom['DATE_TIME'].describe()
```

```python
##Country A only had 19e2e as the unique country. All the products are based on that country and focus can be done there. However, Country_B there were  two unique
##for that, we will find the variations in the products and frequencies to see which product to focus on which country
df_telcom['COUNTRY_B'].groupby(df_telcom['PRODUCT']).describe()
```
